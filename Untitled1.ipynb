```
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sns

from sklearn.linear_model  import LinearRegression

train = pd.read_csv("train1.csv")
test = pd.read_csv("test1.csv")
```

```
train.dtypes #confirm data type
```

```
train["SaleCondition"].unique() #SaleConditionの内容確認
```

```
train.shape[1] #列の要素数_変換前
```

```
#一括で文字列を数値に変換
from sklearn.preprocessing import LabelEncoder

# データタイプがobjectの列の値をラベル化した数値に変換
for i in range(train.shape[1]):
    if train.iloc[:,i].dtypes == object:
        lbl = LabelEncoder()
        lbl.fit(list(train.iloc[:,i].values) + list(test.iloc[:,i].values))
        train.iloc[:,i] = lbl.transform(list(train.iloc[:,i].values))
        test.iloc[:,i] = lbl.transform(list(test.iloc[:,i].values))
```

```
train["SaleCondition"].unique()　#列の要素数_変換後
```

```
#欠損値の確認
train_nan = train.isnull().sum()
train_nan = train_nan[train_nan > 0] 
train_nan
```

```
test_nan = test.isnull().sum()
test_nan = test_nan[test_nan > 0]
test_nan
```

```
train_ID = train["Id"]
test_ID = test["Id"]

y_train = train["SalePrice"]
x_train = train.drop(["Id", "SalePrice"], axis = 1)
x_test = test.drop(["Id"], axis = 1)

Xmat = pd.concat([x_train, x_test]) #concat(["A", "B"]) combine A and B
Xmat = Xmat.drop(["LotFrontage", "MasVnrArea", "GarageYrBlt"], axis = 1)
#欠損値が多いため

Xmat = Xmat.fillna(Xmat.median()) #fill out nan as median

#check where ther are nan or not
Xmat_nan = Xmat.isnull().sum()
Xmat_nan = Xmat_nan[Xmat_nan > 0]
Xmat_nan
```

```
Xmat["TotalSF"] = Xmat["TotalBsmtSF"] + Xmat["1stFlrSF"] + Xmat["2ndFlrSF"]
#conbine totalarea = base + 1 + 2
```

```
sns.distplot(y_train)
plt.show()

# distribution of house price
```

```
#use log so make few scattar
y_train = np.log(y_train)

sns.distplot(y_train)
plt.show()
```

```
#実験とテストのデータを分ける
X_train = Xmat.iloc[:train.shape[0],:]
X_test = Xmat.iloc[train.shape[0]:,:]
```

```
#ランダムフォレストで特徴量の需要量を考える
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 80, max_features = "auto")
rf.fit(X_train, y_train)

"""
np.argsort() ... arange the index following much importnce
"-" AAA → from big importance to small importance
"""
ranking = np.argsort(-rf.feature_importances_)
f, ax = plt.subplots(figsize = (13, 13)) #lager graph
sns.barplot(x = rf.feature_importances_[ranking], y = X_train.columns.values[ranking],
                                        orient = "h")

ax.set_xlabel("feature importance")
plt.tight_layout()
plt.show()
```

```
#use the top 30 feature 
X_train = X_train.iloc[:, ranking[:30]]
X_test = X_test.iloc[:, ranking[:30]]
```

```
X_train = (X_train - X_train.mean())/X_train.std()
X_test = (X_test - X_test.mean())/X_test.std()
#標準化
```

```
#SalePrice/each feature
fig = plt.figure(figsize=(12,7))
for i in np.arange(30):
    ax = fig.add_subplot(5,6, i + 1)  #?????????????
    sns.regplot(x=X_train.iloc[:,i], y=y_train)

plt.tight_layout()
plt.show()       
```

```
y_train = Xmat['SalePrice']
X_train = Xmat.drop(['SalePrice'], axis=1)
```

```
#線回帰分析で予測(LinearRegression)
slr = LinearRegression()

slr.fit(X_train, y_train)

print(slr.coef_[0])
print(slr.intercept_)
```

```
#logを戻す
y_test_pred = np.exp(slr.predict(X_test))
```
